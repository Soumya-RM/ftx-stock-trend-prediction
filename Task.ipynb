## Time-Series Analysis

#### Predict Future Prices

```python
df1=df.copy()
```

```python
df1['Date']=pd.to_datetime(df1['Date'])
df1 = df1.set_index('Date')
df1=df1.sort_values('Date')
```

```python
df1.head()
```

```python
lags = 10  # Use last 10 days as features
for i in range(1, lags + 1):
    df1[f'lag_{i}'] = df1['Closing Amount'].shift(i)
```

```python
df1 = df1.dropna()
```

```python
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop(columns=['Closing Amount'])
y = df1['Closing Amount']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=False)
```

```python
model = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.01)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
future_preds = []
last_known_values = x.iloc[-1].values
```

```python
for _ in range(30):
    next_pred = model.predict(last_known_values.reshape(1, -1))[0]
    future_preds.append(next_pred)
    last_known_values = np.roll(last_known_values, -1)
    last_known_values[-1] = next_pred
```

```python
future_dates = pd.date_range(start=df1.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')
```

```python
forecast_df = pd.DataFrame({'Date': future_dates, 'Predicted Closing Price': future_preds})
forecast_df.set_index('Date', inplace=True)
print(forecast_df)
```

#### Identify Price Anomalies

```python
df6 = df.copy()
```

```python
df6['Date']=pd.to_datetime(df6['Date'])
df6 = df6.set_index('Date')
df6 = df6.sort_index()
```

```python
df6['Moving_Avg'] = df6['Closing Amount'].rolling(window=30).mean()
```

```python
df6['Deviation'] = abs(df6['Closing Amount'] - df6['Moving_Avg'])
df6['Percent_Deviation'] = (df6['Deviation'] / df6['Moving_Avg']) * 100
```

```python
threshold = df6['Percent_Deviation'].std() * 2  
df6['Anomaly'] = df6['Percent_Deviation'] > threshold
```

```python
anomalies = df6[df6['Anomaly'] == True]
print(anomalies[['Closing Amount', 'Moving_Avg', 'Percent_Deviation']])
```

```python
plt.figure(figsize=(12,6))
plt.plot(df6.index, df6['Closing Amount'], label="Closing Price", color='blue', alpha=0.6)
plt.plot(df6.index, df6['Moving_Avg'], label="30-Day Moving Average", color='orange', linestyle='dashed')
plt.scatter(anomalies.index, anomalies['Closing Amount'], color='red', label="Anomalies", marker='o', s=100)
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Price Anomalies Detection")
plt.legend()
plt.show()
```

#### Stock Price Prediction

```python
df7 = df.copy()
```

```python
df7['Date']=pd.to_datetime(df7['Date'])
df7 = df7.set_index('Date')
df7 = df7.sort_index()
```

```python
df7['Prev_Close_1'] = df7['Closing Amount'].shift(1)  
df7['Prev_Close_2'] = df7['Closing Amount'].shift(2)  
df7['Prev_Close_3'] = df7['Closing Amount'].shift(3)  
df7.dropna(inplace=True)  
```

```python
x = df7[['Prev_Close_1', 'Prev_Close_2', 'Prev_Close_3']]
y = df7['Closing Amount']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=False)
```

```python
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
future_dates = pd.date_range(start=df7.index[-1], periods=61, freq='D')[1:]
future_preds = []
last_known = df7.iloc[-1][['Prev_Close_1', 'Prev_Close_2', 'Prev_Close_3']].values
```

```python
for _ in range(60):
    next_pred = model.predict(np.array(last_known).reshape(1, -1))[0]
    future_preds.append(next_pred)
    last_known = [next_pred] + list(last_known[:-1])
```

```python
future_df = pd.DataFrame({'Date': future_dates, 'Predicted_Close': future_preds})
print(future_df.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(df7.index, df7['Closing Amount'], label="Actual Close", color='blue')
plt.plot(future_df['Date'], future_df['Predicted_Close'], label="Predicted Close", color='red', linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Price Prediction (Next 60 Days)")
plt.legend()
plt.show()
```

### Cross-validation(KFold technique)

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
models = {
    "Linear Regression": lr,  
    "Ridge": lr_r,
    "Lasso": lr_l,
    "Decision Tree (Base)": regtree,
    "Decision Tree (Grid Search)": regtree1_cv,
    "Bagging (Base)": bag_reg,
    "Bagging (Grid Search)": regtree_bag_esti,
    "Random Forest (Base)": regtree_rf,
    "Random Forest (Grid Search)": regtree_rf_esti,
    "Gradient Boosting (Base)": regtree_gb,
    "Gradient Boosting (Grid Search)": regtree_gb_esti,
    "XG Boost (Base)": regtree_xgb,
    "XG Boost (Grid Search)": regtree_xgb_esti
}
```

```python
cv_results = {}
for name, model in models.items():
    scores = cross_val_score(model, x, y, cv=cv, scoring='r2')  
    cv_results[name] = (np.mean(scores), np.std(scores))
```

```python
for name, (mean_score, std_dev) in cv_results.items():
    print(f"{name}: Mean R = {mean_score:.4f}, Std = {std_dev:.4f}")
```

```python
df_results = pd.DataFrame(cv_results, index=["Mean R", "Std Dev"]).T
```

```python
df_results = df_results.sort_values(by="Mean R", ascending=False)
```

```python
print(df_results)
```

* After cross-validation there is a slight improvement in the R2 scores for all the models.
* Linear Regresson is the best model.
